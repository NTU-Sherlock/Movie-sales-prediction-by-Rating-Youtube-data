```python
def subs(s):
    return s[28:35]
```

```python
def zero_out(s):
    if s[0] =='0':
        s = s[1:len(s)]
    else:
        s = s
    return s
```

```python
imdb['imdbId'] = imdb.movie_imdb_link.apply(subs)
```

```python
m1 = pd.merge(yt_link,imdb, on = 'imdbId', how = 'inner')
```

```python
m1.shape
```

```python
def num_only(s):
    s = ''.join(c for c in s if c.isdigit())
    return s
```

```python
def substring(s):
    """Remove `start` characters from the beginning and `end` 
    characters from the end of string `s`.

    Examples
    --------
    >>> substring('abcde', 0, 3)
    'abc'
    >>> substring('abcde', 1, None)
    'bcde'
    """
    return s[-4:]
```

```python
m1['year'] = m1['title'].apply(num_only)
```

```python
m1['year'] = m1['year'].apply(substring)
```

```python
m1['year'] = pd.to_numeric(m1.year,errors='coerce',downcast = 'signed')
```

```python
m1_recent = m1[m1['year']>2006]
```

```python
from time import sleep
```

```python
dates = []
views = []
likes = []
dislikes = []
```

```python
a = 1
```

```python
for url in m1_recent.full_link:
    request = requests.get(url)
    content = request.content
    soup = BeautifulSoup(content, "html.parser")
    #Get released date
    try:
        date = soup.select(".watch-time-text")[0].get_text("#").split("#")[0]
        dates.append(''.join(c for c in date if c.isdigit())) 
    except:
        dates.append('NaN')
    #Get the Total view
    try:
        view = soup.select(".watch-view-count")[0].get_text("#").split("#")[0]
        views.append(''.join(c for c in view if c.isdigit()))
    except:
        views.append('NaN')
    #Get the likes
    try:
        like = soup.select(".yt-uix-clickcard ")[3].select('button')[0].get('aria-label')
        likes.append(int(''.join(c for c in like if c.isdigit())))
    except:
        likes.append('NaN')
    #Get the dislikes
    try:
        dislike = soup.select(".yt-uix-clickcard ")[4].select('button')[0].get('aria-label')
        dislikes.append(int(''.join(c for c in dislike if c.isdigit())))
    except:
        dislikes.append('NaN')
    print( 'finished number %d,'%a + url )
    a = a+1
    sleep(5)
```

```python
m1_recent['dates'] = dates
m1_recent['views'] = views
m1_recent['likes'] = likes
m1_recent['dislikes'] = dislikes
```
